```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import os
```

```python
celebs=[]
num_of_pics=[]
files=os.listdir("Bollywood_celeb_face_localized")
for file in files:
    if file==".DS_Store":
        pass
    else:
        pics=os.listdir("Bollywood_celeb_face_localized/"+file)
        celebs.append(file)
        num_of_pics.append(len(pics))

print(celebs)
print(num_of_pics)
```

```python
plt.figure(figsize=(14, 14))
plt.barh(celebs, num_of_pics, color='skyblue')
plt.xlabel('Values')
plt.title('Bar Plot of Names and Corresponding Values')
plt.tight_layout()  # Adjusts plot to fit in the figure area.
plt.show()
```
